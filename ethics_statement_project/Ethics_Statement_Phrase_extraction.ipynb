```python
import pytextrank
import spacy

nlp = spacy.load("en_core_web_sm")
nlp.add_pipe("textrank")
```

```python
from icecream import ic
import pathlib


text = pathlib.Path("ethics_statement_cancer_clinical_trial_15.txt").read_text()
doc = nlp(text)
```

```python
tr = doc._.textrank
ic(tr.elapsed_time);
```

```python
 for phrase in doc._.phrases:
    ic(phrase.count, phrase.rank, phrase.text)
    ic(phrase.chunks)
```

```python
nlp = spacy.load("en_core_web_sm")
nlp.add_pipe("textrank", config={ "stopwords": { "word": ["NOUN"] } })

doc = nlp(text)

for phrase in doc._.phrases[:10]:
    ic(phrase)
```
